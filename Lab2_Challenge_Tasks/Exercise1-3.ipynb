```python
# APPENDIX EXERCISE 1

target_name = "SALAHOP"
SEED_NUM = "0"
redacted_name = ""

for a in target_name:
    if a in ("A", "E", "I", "O", "U") :
        redacted_name = redacted_name = SEED_NUM
    else:
        redacted_name = redacted_name + a

print("Original Surname:", target_name)
print("SEED_NUM used:", SEED_NUM)
print("Redacted Output:", redacted_name)
```

```python
# APPENDIX EXERCISE 2

ENGINEER = "Reynald James"
SEED_NUM = int(SEED_NUM)
loops = SEED_NUM + 3

first_letter = ENGINEER[0]
byte_val = ord(first_letter)

A = byte_val
B = SEED_NUM
C = loops

verification_hash= (A * B) - C

print(f"(A) FIRST LETTER ASCII:", {first_letter})
print(f"(B) SEED NUM:", SEED_NUM)
print(f"(C) Total Loops:", loops)
print("Verification Hash", verification_hash)
```

```python
# APPENDIX EXERCISE 3

correct_password = "rysalahop"
attempts = 0

while True:
    input_attempts = input("Enter the password: ")
    attempts += 1

    if input_attempts == correct_password:
        print("Correct!")
        break
    else: 
        print("Incorrect.")

print(f"Total attempts: {attempts}")
```
